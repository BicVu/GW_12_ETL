```python
import pandas as pd
```

```python
# Read in data
census_csv = "data/Census_States_Clean.csv"
census_states_df = pd.read_csv(census_csv, header=None)
census_states_df.dropna(how="any")
census_states_df.head()
```

```python
# Check if "NaN: is not a number values or string
nan_value = census_states_df[0][1]
print(f"Nan is a {type(nan_value)}")
census_states_df.fillna
census_states_df.head()
```

```python
# Must convert not a number to string to use as an header
census_states_df[0] = census_states_df[0].astype(str)
```

```python
# Set new headers and keep only rows with usable values
new_header = census_states_df.iloc[1] # Grab the first row for the header
census_df = census_states_df[2:] # Take the data less the header row
census_df.columns = new_header # Set the header row as the df header
census_df.head()
```

```python
# Print column names
census_df.columns
```

```python
# Must convert not a number column header to a string rename
#census_states_df[0] = census_states_df[0].astype(str)
```

```python
# Filter only rows with necessary values and rename column header.
# Note headers start out as string and float values.
census_df = census_df[["nan", 2015.0, 2017.0]]
# Rename headers as strings that makes sense
census_df = census_df.rename(columns={"nan":"State", 2015.0:"2015", 2017.0:"2017"})
census_df.head()
```

```python
# Check data types. Convert floats to ints.
print(census_df.dtypes)
census_df["State"] = census_df["State"].astype(str)
census_df["2015"] = census_df["2015"].astype(int)
census_df["2017"] = census_df["2017"].astype(int)
print(census_df.dtypes)
```

```python
# Sort values in alpha order
census_df = census_df.sort_values("State", ascending=True)
```

```python
# Export to csv
census_df.to_csv("output/Census_ETL.csv", index=False, header=True)
```
