```python
import pandas as pd
```

```python
# Read in ETL data from previous scripts that exported to output
census_csv = "output/Census_ETL.csv"
census_etl = pd.read_csv(census_csv)
census_etl.head()
```

```python
lunches_csv = "output/Lunch_ETL.csv"
lunches_etl = pd.read_csv(lunches_csv)
lunches_etl.head()
```

```python
# Merge and rename columns
lunch_census = pd.merge(lunches_etl, census_etl, on="State", suffixes=("_Lunch", "_Census"))
lunch_census.head()
```

```python
# Check data types
print(lunch_census.dtypes)
```

```python
# Calculate percentage of population of each state that receive school lunches
lunch_census['Percentage_2015'] = (lunch_census['2015_Lunch']/lunch_census['2015_Census']*100).round(1)
lunch_census['Percentage_2017'] = (lunch_census['2017_Lunch']/lunch_census['2017_Census']*100).round(1)
lunch_census.head()
```

```python
# Filter for only necessary columns
lunch_census = lunch_census[["State", "Percentage_2015", "Percentage_2017"]]
lunch_census
```

```python
# Export to csv
lunch_census.to_csv("output/Lunch_Census_ETL.csv", index=False, header=True)
```
