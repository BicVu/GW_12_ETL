```python
import pandas as pd
```

```python
# Read in data
lunches_csv = "data/School_Lunch.csv"
lunches_df = pd.read_csv(lunches_csv)
lunches_df.head()
```

```python
# Check data type of "--" value
x = lunches_df.iloc[3, 2]
type(x)
```

```python
# check all data types
lunches_df.dtypes
```

```python
# Obesity chart only has odd years so keep only columns with matching data
lunches_df = lunches_df[["State/Territory", "FY 2015", "FY 2017"]]
# Rename columns to remove spaces
lunches_df = lunches_df.rename(columns={"State/Territory":"State", "FY 2015":"FY_2015", "FY 2017":"FY_2017"})
lunches_df.head(20)
```

```python
# check "--" value again and remove those rows
x = lunches_df.iloc[59, 2]
print(type(x))
lunches_df = lunches_df[lunches_df.FY_2015  != '--']
lunches_df
```

```python
# Rename columns to match other year column names
lunches_df = lunches_df.rename(columns={"State":"State", "FY_2015":"2015", "FY_2017":"2017"})
lunches_df.head(20)
```

```python
print(lunches_df.dtypes)
# numbers still in string format. Remove comas.
lunches_df["2015"] = lunches_df["2015"].str.replace(",", "")
lunches_df["2017"] = lunches_df["2017"].str.replace(",", "")
lunches_df["2015"] = pd.to_numeric(lunches_df["2015"])
lunches_df["2017"] = pd.to_numeric(lunches_df["2017"])

print(lunches_df.dtypes)
# lunches_df
```

```python
# Export to csv
lunches_df.to_csv("output/Lunch_ETL.csv", index=False, header=True)
```
