```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import psycopg2
from sqlalchemy import create_engine
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy_utils import create_database, database_exists, drop_database
```

```python
# Read in cleaned data from previous scripts
csv = "output/Obesity_ETL.csv"
obesity_df = pd.read_csv(csv)
obesity_df.head()
```

```python
# Read in cleaned data from previous scripts
csv = "data/Lunch_Census_ETL.csv"
lunch_df = pd.read_csv(csv)
lunch_df.head()
```

```python
# Check data types to make sure they match
print(obesity_df.dtypes)
print(lunch_df.dtypes)
```

```python
# Merge and rename columns
# Inner join will only keep states and territories with matching info
compare_df = pd.merge(obesity_df, lunch_df, on="State", suffixes=("_Obesity", "_Lunch"))
compare_df.head()
```

```python
# Export to csv
compare_df.to_csv("output/Comparison.csv", index=False, header=True)
```

```python
# Save output to database
# create db connection
connection_string = "postgres:postgres@localhost:5432/ETL"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

```python
connection = engine.connect()
Base = declarative_base()
Base.metadata.drop_all(engine)
```
